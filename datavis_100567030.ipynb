```python
x = np.linspace(0,2,100)
plt.figure(figsize=(8,6))
plt.plot(x, x, label="linear", color="#43b2a1", linewidth=1.5, linestyle="-.")
plt.plot(x, np.cos(x), label='cosine', color="red", linewidth=2, linestyle=":")
plt.plot(x, np.sin(x), label="sine", color="purple")
plt.xlabel("x label", size=13)
plt.ylabel("y label", size=13)
plt.title("Pyplot simple implementation", size=17)
plt.legend(loc="lower right")
```
